```python
import matplotlib.pyplot as plt
import numpy as np

#Definerer funksjonen
def f(x):
    return -x**2-5

#Skriver ut x - verdier i intervallet -10 - 10 med 200 punkts tetthet
x = np.linspace(-10,10,200)
y = f(x)

#Tittel på x- akse
plt.xlabel("x - verdier")

#Tittel på y- akse
plt.ylabel("y - verdier")

#Legger til forklaring
plt.legend("f")

#Tittel på plottet
plt.title("En graf")

#Legger til rutenett
plt.grid()

#Formatering av grafen
#plt.plot(x,y,  ":k")
#plt.plot(x,y,  "--b")
plt.plot(x,y, "c")

#Skriver grafen til skjerm
plt.show()
```
